```python
aggs=[]
for l in links:
  if l.find('100')==-1:
    d=20.0
  else:
    d=100.0
  df= pd.read_csv(l)
  print(l[9:-4],d)
  df.client_account_id= df.client_account_id.astype('str')
  sf= df.agg(met_dict)
  sf['vends_per_mc']= round(sf.vend_cnt/ sf.machines,2)
  sf['vends_per_txntr']= round(sf.vend_cnt/sf.uniq_transactors, 2)
  sf['dollar_vend_per_txntr']= round(sf.amt/ (100.00*sf.uniq_transactors),2)
  sf['dollar_vend_per_mc']= round(sf.amt/ (100.0*sf.machines),2)
  sf['flag']= l[9:-4]
  sf['amt']= sf.amt/(100.0*d*sf.client_account_id)
  sf['vend_cnt']= sf.vend_cnt/(d*sf.client_account_id)
  sf['machines']= sf.machines/(d)
  sf['uniq_transactors']= sf.uniq_transactors/(d*sf.client_account_id)
  aggs.append(sf)
```
