```python
import pandas as pd
```

```python
ef= pd.read_csv('/content/DFORE34events.csv')
```

```python
af= pd.read_csv('/content/DFOREagg.csv')
```

```python
set(ef.client_account_id).difference(af.client_account_id)
```

```python
df= af.merge(ef, how='inner',on= 'client_account_id', suffixes=('_agg','_events'))
```

```python
df['vends_delta']= df.vends_events- df.vends_agg
df['vend_dollar_delta']= df.vend_dollar_events- df.vend_dollar_agg
df['users_delta']= df.users_events- df.users_agg
```

```python
df.to_csv('DFORE34deltas.csv')
```
