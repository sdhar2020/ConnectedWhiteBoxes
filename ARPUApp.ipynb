```python
import pandas as pd
```

```python
df= pd.read_csv('/content/arpua_app.csv')
```

```python
df.head()
```

```python
met_dict= {'user_id': 'nunique', 'amt_fw4_weeks': 'sum', 'amt_fw8_weeks': 'sum', 'amt_bc4_weeks': 'sum', 'amt_bc8_weeks': 'sum'}
```

```python
df1= df.groupby(['eventName']).agg(met_dict)
```

```python
df1
```

```python
emails=['emailSend', 'emailOpen', 'emailClick']
```

```python
lst= ['amt_bc8_weeks', 'amt_bc4_weeks', 'amt_fw4_weeks', 'amt_fw8_weeks']
```

```python
df1.divide(df1.iloc[:,0], axis=0).iloc[:,1:].loc[emails, lst]/100.00
```

```python
mf= pd.melt(df, id_vars=['user_id', 'eventName'], value_vars=['amt_fw4_weeks', 'amt_fw8_weeks', 'amt_bc4_weeks', 'amt_bc8_weeks'], value_name="Value")
```

```python
map_dict= {'amt_fw4_weeks':4, 'amt_fw8_weeks':8, 'amt_bc4_weeks':-4, 'amt_bc8_weeks':-8}
```

```python
mf['period']= mf.variable.map(map_dict)
```
