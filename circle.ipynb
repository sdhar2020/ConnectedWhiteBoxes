```python
import pandas as pd
```

```python
df= pd.read_excel('Simpli.fi CSC GO GO GO  targets Dec to  Feb.xlsx', index_col=0)
```

```python
import numpy as np
import json
import geog
import shapely.geometry

def find_circle(long, lat, num_pnts=25, radius=10):
  p = shapely.geometry.Point([long, lat])
  n_points = num_pnts
  d = radius  # meters
  angles = np.linspace(0, 360, n_points)
  polygon = geog.propagate(p, angles, d)
  circle= json.dumps(shapely.geometry.mapping(shapely.geometry.Polygon(polygon)))
  return(circle)
```

```python
circle= find_circle(-90.0667, 29.9500)
```

```python
df['circle']= df.apply( lambda x: find_circle(x.Longitude, x.Latitude), axis=1)
```
