```python
import pandas as pd
```

```python
df= pd.read_csv('/content/DFORE129D2.zip', compression='zip', low_memory= False)
```

```python
df['BaseCycleBkt']= pd.cut(x=df.BaseCycle, bins=[-1, 0, 1, 10000000],
                    labels=['Zeros', 'Ones', 'MoreThanOne'])
 
```

```python
df.BaseCycleBkt.value_counts()
```

```python
df['TopOffsBkt']= pd.cut(x=df.TopOffs, bins=[-1, 0, 1, 9,99,999,9999, 99999999999999999],
                    labels=['Zeros', 'Ones', 'Units', 'Tens', 'Hundreds', 'Thousands', 'EvenMore000'])
```

```python
df['payment_channel']= df.payment_type.map({'Coin': 'Cash', 'bill': 'Cash', 
                                            'Apple-Pay': 'MicroTxn', 
                                            'Google-Pay': 'MicroTxn', 
                                            'CreditCard-SR': 'MicroTxn', 
                                            'Stored-Value': 'Stored-Value'})
```

```python
pd.crosstab(df.payment_channel, df.source, margins=True)
```

```python
pd.crosstab(df.BaseCycleBkt, df.TopOffsBkt, margins=True)
```

```python
df1= df.loc[df.payment_channel== 'Stored-Value']
pd.crosstab(df1.BaseCycleBkt, df1.TopOffsBkt, margins= True)
```

```python
df1= df.loc[df.payment_channel== 'Cash']
pd.crosstab(df1.BaseCycleBkt, df1.TopOffsBkt)
```

```python
df1= df.loc[df.payment_channel== 'MicroTxn']
pd.crosstab(df1.BaseCycleBkt, df1.TopOffsBkt, margins= True)
```
