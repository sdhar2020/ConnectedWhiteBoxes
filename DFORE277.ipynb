```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('/content/DFORE277.csv')
```

```python
df.columns= ['Date', 'Participants', 'ParticipantVends', 'Delivered', 'DelveredVends']
```

```python
df['Date']= pd.to_datetime(df.Date)
```

```python
df['DoW']=df.Date.dt.dayofweek
```

```python
df['Month']= df.Date.dt.month_name()
```

```python
df=df.loc[~(df.Date== '2021-10-31')]
```

```python
df=df.loc[~(df.Date== '2021-10-30')]
```

```python
df['NonParticipants']= df.Delivered- df.Participants
```

```python
df['NonParticipantsVends']= df.DelveredVends- df.ParticipantVends
```

```python
df.head()
```

```python
df['Pvendsperuser']= df.ParticipantVends/df.Participants
df['NPvendsperuser']= df.NonParticipantsVends/ df.NonParticipants
df['Allvendperuser']= df.DelveredVends/ df.Delivered
```

```python
df.groupby('Month')['Allvendperuser', 'Pvendsperuser', 'NPvendsperuser'].agg(np.mean)
```

```python
df.head()
```

```python
df5=df.groupby(['Month', 'DoW']).agg({'Pvendsperuser': np.mean, 'NPvendsperuser': np.mean, 'ParticipantVends': np.sum,
                                  'NonParticipantsVends': np.sum}).reset_index()
```

```python
df5['ParticipantVends%']= df5['ParticipantVends']/(df5['ParticipantVends']+ df5['NonParticipantsVends'])
df5['NonParticipantsVends%']= df5['NonParticipantsVends']/(df5['ParticipantVends']+ df5['NonParticipantsVends'])
```

```python
df5.plot(x= 'DoW', y= ['NonParticipantsVends%', 'ParticipantVends%'], stacked= True, kind= 'bar')
```

```python
df2= pd.read_csv('/content/DFORE277a.csv')
```

```python
df2['DeltaNovOct']= df2.NovVends- df2.OctVends
```

```python
rng=df2.DeltaNovOct.describe(percentiles=[.05, .95])[['5%', '95%']].values
```

```python
df4=df2.loc[df2.DeltaNovOct.between(rng[0], rng[1])]
```

```python
df4.OctVends.describe()
```

```python
pd.cut(df4.DeltaNovOct, bins= np.linspace(min(df4.DeltaNovOct), max(df4.DeltaNovOct), 10), 
       labels=[1,2,3,4,5,6,7,8,9]).value_counts()
```

```python
bin_labels_5 = ['OctQ1', 'OctQ2', 'OctQ3', 'OctQ4', 'OctQ5']
df4['OctVendsQ'] = pd.qcut(df4['OctVends'],
                              q=[0, .2, .4, .6, .8, 1])
```
