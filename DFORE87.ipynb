```python
def change_val(df, colname):
  colFlg= colname+'Flg'
  df[colFlg]= None
  df.loc[(df[colname] < 0), colFlg]= 'Decreased'
  df.loc[(df[colname] == 0), colFlg]= 'No Change'
  df.loc[(df[colname] > 0), colFlg]= 'Increased'
  return(df)
```

```python
delist=['deltaAllVends', 'deltaMicroVends', 'deltaCashVends', 'deltaStoreValVends'] 
```

```python
for d in delist:
  change_val(df1, d)
```
