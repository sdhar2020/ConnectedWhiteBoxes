```python
dfs=[]
for sheet in ['control', 'test1', 'test2']:
  df= pd.read_excel(lnk, sheet_name= sheet)
  df['cohort']= sheet
  dfs.append(df)
```

```python
cutoffs={1: [.01,.99],
         2: [.05, .95],
         3: [.1,.9]}
```

```python
met_dict= {'as400id': 'nunique', 'total_transacted_days_pre': 'sum', 'machine_cnt_pre': 'sum',
       'total_vends_pre': 'sum', 'total_amount_pre': 'sum', 'Days Diff.': 'sum', 'Machine Diff.': 'sum',
       'Vends Diff.': 'sum', 'Amount Diff.': 'sum',  'total_transacted_days_post': 'sum',
       'machine_cnt_post': 'sum', 'total_vends_post': 'sum', 'total_amount_post': 'sum', 'DDCut': 'min', 'MDCut': 'max'}
```

```python
aggs=[]
for t in [.96, .95, .92, .9]:
  df= pd.concat(dfs, axis=0)
  # df['McDayDiff']= df['Days Diff.'].abs()*df['Machine Diff.'].abs()
  dq= df.groupby('cohort')['Days Diff.', 'Machine Diff.'].agg(lambda x: x.abs().quantile(t)).rename(columns={'Days Diff.': 'DDCut','Machine Diff.': 'MDCut' })
  df= df.merge(dq, left_on='cohort', right_on='cohort')
  df1= df.loc[(df['Days Diff.'].abs() <= df.DDCut) & (df['Machine Diff.'].abs() <= df.MDCut)]
  df2= df1.groupby('cohort').agg(met_dict).reset_index()
  df2['CutOffPercentile']= t
  # df2['HighCutOff']= v[1]
  aggs.append(df2)
```

```python
met_dict= {'as400id': 'nunique', 'total_transacted_days_pre': 'sum', 'machine_cnt_pre': 'sum',
       'total_vends_pre': 'sum', 'total_amount_pre': 'sum', 'Days Diff.': 'sum', 'Machine Diff.': 'sum',
       'Vends Diff.': 'sum', 'Amount Diff.': 'sum',  'total_transacted_days_post': 'sum',
       'machine_cnt_post': 'sum', 'total_vends_post': 'sum', 'total_amount_post': 'sum',
       'McDayDiff': ['min', 'max'], 'qlow': 'min', 'qHi': 'max'}
```

```python
aggs=[]
for i,v in cutoffs.items():
  df= pd.concat(dfs, axis=0)
  df['McDayDiff']= df['Days Diff.']*df['Machine Diff.']
  dq= df.groupby('cohort')['McDayDiff'].agg([('qlow', lambda x: x.quantile(v[0])), ('qHi', lambda x: x.quantile(v[1]))]).reset_index()
  df= df.merge(dq, left_on='cohort', right_on='cohort')
  df1= df.loc[df.McDayDiff.between(df.qlow, df.qHi)]
  df2= df1.groupby('cohort').agg(met_dict).reset_index()
  df2['LowCutOff']=v[0]
  df2['HighCutOff']= v[1]
  aggs.append(df2)
```
