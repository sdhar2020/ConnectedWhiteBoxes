```python
import pandas as pd
```

```python
df= pd.read_csv('/content/DFORE82sprdsht.csv')
```

```python
amts= ['amt_fw2_weeks',	'amt_fw4_weeks',	'amt_bc2_weeks',	'amt_bc4_weeks'	]
```

```python
for a in amts:
  df['act_'+a[4:]]= 'False'
  df['act_'+a[4:]]= (df[a]!=0)
```

```python
amt_sum= {'amt_fw2_weeks':'sum',	'amt_fw4_weeks': 'sum',	'amt_bc2_weeks': 'sum',	
          'amt_bc4_weeks': 'sum','act_fw2_weeks':'sum',	'act_fw4_weeks':'sum', 'act_bc2_weeks': 'sum',	'act_bc4_weeks': 'sum',
          'vnd_fw2_weeks': 'sum', 'vnd_fw4_weeks': 'sum',	'vnd_bc2_weeks': 'sum',	'vnd_bc4_weeks': 'sum'}
```

```python
df.loc[df.eventName!='emailBounce'].groupby('eventName').agg(amt_sum)
```
